# CONTAR LOS CARACTERES 

```python
import re
```

LEE EL ARCHIVO 

```python
def read(file):
    with open(f"books/{file}", encoding = 'utf-8') as text:
        return text.readlines()
```

CUENTA LOS CARACTERES QUE SE REPITEN EN UNA LINEA Y SI NO EXISTEN LOS AGREGA AL DICCIONARIO, ASI COMO TAMBIEN CUENTA EL TOTAL DE CARACTERES EN LA LINEA

```python
def counter(lines):
    start = False
    characters = {}
    total_chars = 0
    for line in lines:                      #* HERE WE ITERATE THE LINES IN THE BOOK

        x = re.search('^\*{3}\s{0,3}START|^\*{3}\s{0,3}END', line)      #* HERE WE FIND THE START OF THE BOOK WITH A REGEX THAT CHECK IF THE LINE STARTS WITH 3 ASTERISKS *
        if x is not None and start is False:    #* WHEN WE FIND THE LINE WICH STARTS WIT 3 ASTERISK AND THE SATART FLAG IS FALSE 
            print(line)                     #* PRINT THAT LINE
            start = True                    #* AND NOW THE FLAG START IS TRUE AND WE CAN START CHEKING AND PROCESSING THE NEXT LINES
            x = None                        #* RESTART OUR INDICATOR OF THE FIRST LINE BECAUSE WE WILL USE TO FIND THE END OF THE BOOK
            continue                        #* JUMP TO THE NEXT LINE IN LINES (NEXT ITERATION) 
        if x is not None and start is True:     #* IF WE FOUND THE LAST LINE OF THE BOOK AND WE HAVE ALREADY STARTED 
            print(line)                     #* PRINT THE LAST LINE AND GET OUT THE CICLE FOR
            break

        if start is True:
            for char in line:   
                if char in characters:      
                    characters[char] += 1            #* WE CHECK EVERY CHAR IN THE LINE AND 
                if char not in characters:  #* CHECK IF THE CHAR IS ALREADY REGISTERED IN THE DICTIONARY
                    characters[char] = 1    #* IF THE CHAR IS NOT REGISTERED IN THE DICTIONARY WE INITIALIZE AS KEY WITH VALUE = 1

                total_chars += 1

    print(f"total chars {total_chars}")
    return [characters, total_chars]
```

PRESENTA LOS RESULTADOS OBTENIDOS

```python
def result(dictionary):
    characters = dictionary[0]
    total = dictionary[1]
    if total == 0:
        total = 1
    for char in enumerate(sorted(characters)):         #* HERE WE ITERATE THE SORTED DICTIONARY 
    
        if char[1] == "\n":                    #* IF THE CHAR IS A LINE FEED WE PUT AN SPECIAL PRINT
            print(f'{char[0]}.- Line Feed \ n has been used "{characters[char[1]]}" \t  "{characters[char[1]]/total}%"')
            continue
        
        print(f'{char[0]}.- this char "{char[1]}" has been repeated "{characters[char[1]]}" \t "{characters[char[1]] * 100 /total}%"') #*  DEPLOYS THE AMOUNT OF TIMES THAT THE CHARACTER HAS BEN FOUND
    print('\n\n')
       
    
```

HACE UNA LIMPIEZA DEL DOCUMENTO QUITANDO TODOS LOS CARACTERES NO ALFANUMERICOS Y TRADUCIENDO CARACTERES ESPECIALES A ASCII



```python
def clean(line):
    line = line.rstrip()                                #* HERE WE TAKE OUT THE \n
    line = line.lower() 

    line = re.sub('\W', "", line)                   #* HERE WITH THE REGEX CHANGE THE NON ALFANUMERIC CHARACTERS BY NONE
     #* AND THEN WE REPLACE EVERY SPECIAL CHARACTER FOR AN ASCII CHARACTER 
    line = re.sub( '[ÿý]', 'y', line)
    line = re.sub( '[ç]', 's', line)
    line = re.sub( '[æ]', 'ae', line)
    line = re.sub( '[ß]', 'ss', line)
    line = re.sub('[âááäāãàå]', 'a',line)
    line = re.sub('[èêééëē]', 'e',line)
    line = re.sub('[îíïīì]', 'i',line)
    line = re.sub('[ôóöōõòø]', 'o',line)
    line = re.sub('[ûúüūù]', 'u',line)

   
    return line;
```

CUENTA LOS CARACTERES QUE SE REPITEN UTILIZANDO LA LIMPIEZA ANTES DE CONTARLOS, TAMBIEN OBTIENE EL NUMERO TOTAL DE CARACTERES

```python
def clean_counter(lines):
    start = False
    characters = {}
    total_chars = 0
    for line in lines:
                                       
        x = re.search('^\*{3}\s{0,3}START|^\*{3}\s{0,3}END', line)
        if x is not None and start is False:
            print(line)
            start = True
            x = None
            continue
        
        if x is not None and start is True:
            print(line)
            break
            
        if start is True:
            line = clean(line)
            for char in line: 
                if char in characters:      
                    characters[char] += 1              #* WE CHECK EVERY CHAR IN THE LINE AND 
                if char not in characters:  #* CHECK IF THE CHAR IS ALREADY REGISTERED IN THE DICTIONARY
                    characters[char] = 1    #* IF THE CHAR IS NOT REGISTERED IN THE DICTIONARY WE INITIALIZE AS KEY WITH VALUE = 1
                
                total_chars += 1
                        
    print(f"total chars {total_chars}")
    return [characters, total_chars]
```

```python
%%time
lines_en = read('en-10012-8_UTF8.txt')
lines_de = read('de-10223-8_UTF8.txt')
lines_fr = read('fr-10682-8_UTF8.txt')
lines_es = read('fr-10682-8_UTF8.txt')

files = [lines_en, lines_de, lines_fr, lines_es]

for file in files:

    get_dict = clean_counter(file)
    result(get_dict)
```
